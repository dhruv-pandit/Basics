## Discrete Random Variables
### Expected Values and Variance
Expected value of random value is theoretical mean of random variable, it is a parameter **not a statistic**
- $ E(x) = \sum xp(x) $ ; $E[g(x)] = \sum g(x)p(x)$
- g(x) can be any function, $x^2$ for example
#### Example:
Suppose 60% of people support the prime minister. If we randomly 2 people
Let X represent the number of people who support
- X 0    1    2
- p 0.16 0.48 0.36
- then $ E(X) = \sum xp(x) = 0*0.16+1*0.48+2*0.36 = 1.2$
- This is the theoretical mean, not one of the values of X. 

### Bernouli Distribution
- Single Trial
- Two outomces (success/ failure)
- P(s) = p; P(F) = 1-p
- $ P(X=x) = p^x(1-p)^{1-x}$ for x = 0,1
- $\mu = p, \sigma^2 = p(1-p)$
#### Example:
- Approx 1 in 200 adults lawyers; one is selected randomly, what is distribution of lawyers?
- Here $p = 1/200$
- If X = number of lawyers in sample:
- $ P(X=x) = (\frac{1}{200})^x(1- \frac{1}{200})^{1-x}$ ; x = 0,1 (either one lawyer or none)
### Binomial Distribution
- $\begin{pmatrix} n \\ x \end{pmatrix} = \frac{n!}{x!(n-x)!}$
- A coin is flipped 100; what is prob heads comes up 60
- You buy lottery ticket once a week for 4 weeks, what is probability oyu win a cash prize exactly twice
  
* N independent trials
* Each trials can have two outcomes (S or F)
* P(S) = p, P(F) = 1-p; X is count of successes in n trials
* $ P(X=x) = \begin{pmatrix} n \\ x \end{pmatrix} p^x(1-p)^{n-x}$ for x = 0,1,...n
* $\mu = E(X) = np ; \sigma^2 = np(1-p)$
#### Example:
A balanced six sided die is rolled 3 times, what is probability 5 comes up exactly twice:
- S = Rolling 5, F = Anything else
- X = num 5 in 3 rolls, binomial distribution n = 3 ; p = 1/6
- $P(X=2) = \begin{pmatrix} 3 \\ 2 \end{pmatrix} (1/6)^2(1-1/6)^{3-2} = 0.0694$
- X~B(n,p)
### HyperGeometric Distribution
- Randomly sampling n objects withjour replacment from asource containing a successe and N-a failures
- X is number of successes
- $ P(X=x) = \frac{\begin{pmatrix} a \\ x \end{pmatrix}\begin{pmatrix} N-a \\ n-x \end{pmatrix}}{\begin{pmatrix} N \\ n \end{pmatrix}}$
- For x = max(0, n+a-N), .. Min(a, n)
- $\mu = n(a/N)$
#### Example:
Suppose a large high school has 1100 female students and 900 male. Sample of 10 selected, what is probability exactly 7 are female?
- Not binomial, because if boss asks you to pick 10, you cant come back with 2 and say you have picked one 7 times
- $ P(X=7) = \frac{\begin{pmatrix} 1100 \\ 7 \end{pmatrix}\begin{pmatrix} 900 \\ 3 \end{pmatrix}}{\begin{pmatrix} 2000 \\ 10\end{pmatrix}}$ = 0.166490
- N = 2000, n = 10, a = 1100
- If binomial, then P(X = 7) = 0.166478
**BD is reasonable approx to the hypergeometric, if not sampling 5% of population then BD is reasonable!**
- 10/2000 = 0.5% so it is reasonable!
### Geometric Distribution
- Distribution of the num of trials needed to get fist success in repeated independent bernoulli trails
- ind. trails
- Two outcomes, S/F and X is the num trails to get the first success
- $ P(X=x) = (1-p)^{x-1}p $ for x = 1, 2, 3...
- $\mu = 1/p, \sigma^2 = 1-p / p^2$

#### Example:
In a large pop. 30% of adults have received CPR training. What is probability that the 6th person sampled is the first that has received training? :
- p = 0.3
- $ P(X = 6) =(1-0.3)^5 \times 0.3 = 0.0504$

### Poisson Distribution
- Events are occurring independently
- Prob that an event occurs in a given length of time does not change
  - Events are random and independent
- X  = number of events in time, has poisson distribution
- $P(X = x) = \frac{\lambda^x e^{-\lambda}}{x!}$
- $ \mu = \lambda ; \sigma^2 = \lambda$

X = number of decays in a 2 second period

* PD with $\mu = np$ approx the BD if n is large and p is small
* Above problem is BD but is approx by Poisson

### Negative Binomial

- $\mu = r/p ; \sigma^2 = r(1-p) / p^2

### Multinomial Distribution

- K outcomes occurr with probabilities p1 .. pk

## Continous Random Variables
### Basics
- RV can take on infinite number of values in an interval, (3 , 5)
- Height can be a CRV, and have a continous probability distribution
- *f(x) = probability density function*
- Probabilities are areas under curve
- $ E(X) = \mu = \int^\infty_\infty xf(x)dx$
- $\sigma^2 = E[(X-\mu)^2] = \int^\infty_\infty (x-\mu)^2f(x)dx = E(X^2) - [E(X)]^2$
  
### Uniform Distribution
- f(x) is constant over possible values of x
- f(x) = 1/d-c between c and d, and 0 elsewhere
- median = $\mu$ = c+d/2
- $\sigma^2 = 1/12 (d-c)^2$

### Normal/Gaussian Distribution
- $X~N(\mu, \sigma^2)$
Std. Normal Distribution is :
- Z~N(0,1)

* If $X~N(\mu, \sigma^2)$

- $Z^2$ has a $\chi^2$ with 1 DOF Where Z~N(0,1)
- $\sum Z_k^2 = \chi^2$ with k DOF

# Sampling Distributions
Let $X_1, X_2, .. X_n$ be n independent observations with mean $\mu$ and variance $\sigma^2$. Then:
- Mean of the sampling distribution of the sample mea is equal to the population mean:
  - $E(\bar{X}) = \mu_{\bar{X}} = \mu$
- Standard deviation of the sampling distribution of \bar{X} is 
  - $\sigma_{\bar{X}} = \sigma / \sqrt{n}$

## Central Limit Theorem
- The sample mean will be approx normally distributed for large sample sizes regardless of the distribution from which we are sampling
If the population is normally distr. then $\bar{X}$ is also normally distri.
- The distribution of the sample mean tends towards normal as the sample size increases
- (VERY ROUGH GUIDELINE) Sample mean is approx is normally distributed if sample size is at least 30 
- $\frac{\bar{X} - \mu}{\sigma / \sqrt{n}} \rightarrow N(0,1) as n \rightarrow \infty$
### Example
- Suppose salaries at company have mean of 62000 and SD of 32000
- If employee selected what is probability that salary exceeds 66000
- We **cannot** use normal distribution to answer this and lookup P(Z > X - mu / sigma) because the salaries are not normally distributed and so we would be wrong
- If 100 employees are selected, then:
  - Then P(\bar{X} > 66000)?
  - Well \bar{X} is going to be normally distributed!
  - So it is P(Z > 66000 - 62000 / 32000 / sqrt 100) = 0.106
  - So we can find out the answer even thought we do not know the distribution of the salaries
## Sampling Distribution
### Intro
SD of a statistic is the PD of the statistic, or the distribution fo the statistic if we were to repeatedly draw samples from the population
### SD Of Sample Mean
TO standardise:
- For a single value:
  - Z = X - $\mu$ / $\sigma$
- For mean of n observations:
  - Z = $\frac{\bar{X} - \mu} {\sigma / \sqrt{n}}$
#### Example
Amount of protein in quarter pound of beef 113.4 is approx normal distributed with mean of 21.4 and sd of 1.9g
- What is probability that randomly selected patty has at least 23.0 g of protein:
  - P(X >= 23) = p(Z >= 23 - 21.4 / 1.9) = P(Z >= 0.842)
- What is probability that 4 randomly selected patties has at least 23.0 g of protein:
  - Now SD = 1.9 / sqrt(4)
  - P(X>= 23) = P(Z >= 23 - 21.4 / 1.9/sqrt(4)) = P(Z>= 1.684)
### Sampling Distribution of the Sample Proportion
- In a random sample of 200 parts from a manufacturing process 18 had major manufacturing defects:
- $\hat{p}$ = 18/200 = 0.09
- Use $\hat{p}$ to estimate p
- $\hat{p}$ = X / n where X is RV with nummber of individuals in sample with the charactersitic and n is number of individuals in sample
  - X is a binomial random variable with parameters n and p
  - $\hat{p}$ takes on n+1 values 0 , 1/n, 2/n ... n
  - $E(\hat{p})$ = E(X/n) = p
    - So $\hat{p}$ is an unbiased estimator of p
  - Var($\hat{p}$) = Var(X/n) = p(1-p) / n
  - Sampling distribution of $\hat{p}$ is approx normal if sample size is large
  - If p is close to 0.5 then sample size can be small to approx normal
  - But if p is close to 0 or 1, then sample size needs to be larger
  - Very Rough : SD of $\hat{p}$ is approx normal if np>= 15 and n(1-p) >= 15
For large sample sizes, sampling distribution of $\hat{p}$ is approx:
N(p, p(1-p)/n)

# Hypothesis Testing
## Z Tests for One Mean
**$\sigma$ is known**
- Testing null hypothesis $H_0  : \mu = \mu_0$
- Against:
  - $H_a  : \mu < \mu_0$ - One Sided 
  - $H_a  : \mu > \mu_0$ - One Sided
  - $H_a  : \mu \ne \mu_0$ - Two Sided
- To test, we use the test statistic :
  - Z = \frac{\bar{X} - \mu_0}{\sigma_{\bar{X}}} = \frac{\bar{X} - \mu_0}{\sigma/\sqrt{n}}
### Example
Tuna contains 0.4 ppm mercury on average. but chef thinks its higher:
- Test $H_0  : \mu = 0.4$
- Against  $H_a  : \mu > 0.4$
- In random sample, average was found to be 0.74, and sigma was 0.08, so is it enough evidence?
- So : Z = 0.74 - 0.40 / 0.08 / sqrt(16) = 17
- P(Z >= 17) = very small, evidence is overwhelming
### Rejection Region Approach
How far out do we need to be to reject hypothesis?
- Choose a value for $\alpha$ = significance level of the test
- Appropriate rejection region, if test falls in region we reject null hypothesis
### P Value Approach
P value is the probability of getting hte observed value of the test statistic or a value with even greater evidence against H_0 if null is true

- p value = P(Z >= value obtained) or P(Z <= value obtained) or 2 * P(Z>= |value obtained|)
- Depends on the observed statistic and alternative hypothesis
- Reject the null hypothesis if for a significance level of $\alpha$:
  - If $p value \le \alpha$
#### Example
Suppose $\mu = 10$, we want to test $H_a$ that it is not equal to 10 for $\alpha = 0.05$
- Z = -2.12, p value = 0.034 so we can evidence against hypothesis significant so we can reject it


## One Sided Vs Two Sided
#### Examples
- Company claims that product has no more that 2 grams of saturated fat:
  - $H_0 : \mu = 2 ; H_a : \mu  > 2$ (one sided)
- Production Process creates wafers with thickness 725 mm, you want to test whether the mean thickness differs
  - $H_0 : \mu = 725 ; H_a : \mu  \ne 725$ (two sided)

### Errors
Type I error is rejecting null hypothesis when it is true, and Type II error is failing to reject it when it is false
- Probability of Type I error given null hypothesis is true is called significance level of the test :
  - P(Type I error | H_0 is true) = $\alpha$
- Power of a test is probability of rejecting null hypothesis given it is false:
  - Power = 1 - $\beta$ = 1- P(Type II error)
#### Statistical Significance VS Practical Significance
Suppose call centre claimes avg wait time is 30 seconds
- $H_0 : \mu = 30 ; H_a : \mu  > 30$
- $\bar{X}$ = 30.6, p value is 0.002
- Strong evidence to reject H_0, but the difference is only 6 seconds so the practical implication might be minimal
- Statistical Significance is related to sample size, so if the sample size is large, so tiny differences will be statistical significant
### Confidence Interval and Hypothesis Testing
Confidence interval is made up of all values of \mu_0 for which we would not reject H_0


## T Tests
- Here we do not know the standard deviation
- *t* = $\frac{\bar{X} - \mu_0} {SE(\bar{X})} = \frac{\bar{X} - \mu_0} {s / \sqrt{n}}$
- Where SE is the estimated standard deviation of the sampling distribution of the sample mean
- *t* does not have normal distribution but *t* distribution with n-1 degrees of freedom
### Example
Does mild dehydration affect reaction times?
Suppose it is known that young women have avg reaction time of 0.95 seconds on a test. For 25 dehydrated women, the avg reaction time of 1 seconds has sd of 0.18 seconds, does this yield strong evidence?
- $H_a : \mu \ne 0.95; H_0:  \mu = 0.95 ; \alpha = 0.05$
- $\bar{X} = 1, s = 0.18, n = 25$
- t = 1 - 0.95 / 0.18/sqrt(n) = 1.389
- DOF = 24, p value is 2 * 0.0888 = 0.1776, so not strong evidence so we do not reject


```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import math
#---Generating Random Data
np.random.seed(6)
pop_ages_1 = stats.poisson.rvs(18, 35, 150000)
pop_ages_2 = stats.poisson.rvs(18, 10, 100000)
pop_ages = np.concatenate((pop_ages_1, pop_ages_2))

min_ages_1 = stats.poisson.rvs(18, 30, 30) # we already know that mu is smaller
min_ages_2 = stats.poisson.rvs(18, 10, 20)
min_ages = np.concatenate((min_ages_1, min_ages_2))
print(pop_ages.mean())
print(min_ages.mean())
 #-- T test
```

```python
stats.ttest_1samp(a = min_ages, popmean=pop_ages.mean())
```

```python
sigma = min_ages.std() / math.sqrt(50)
stats.t.interval(0.95, df = 49, loc = min_ages.mean(), scale = sigma)
#gives us our confience interval
```

## Some More Distributions
### Exponential Distribution
- Predict amount of waiting time until next event
- X ~ Exp(\lambda) where \lambda is not the time duration but event rate (number of character per unit time)
- \lambda is same as in Poisson process
- the probability distribution of the time *between* the events in a Poisson process.
- If the number of events per unit time follows a Poisson distribution, then the amount of time between events follows the exponential distribution.
### Gamma Distribution
- the Gamma Distribution gives us the probability of the waiting time until the nth event.
- $PDF = \frac{\lambda e^{\lambda t}(\lambda t)^{n-1}}{(n-1)!}$
- $\gamma(n) = (n-1)!$
- $PDF = \frac{\lambda e^{\lambda t}(\lambda t)^{n-1}}{\gamma(n)}$
So, the Gamma Distribution is parametrised by two values:
* λ, The rate of events
* n, The number of the events you are waiting for
* The expected waiting time between events is equal to 1/λ
* as λ gets smaller the expected time between events increases

```python
# import packages
import numpy as np
import scipy.stats as stats 
import matplotlib.pyplot as plt# get distributions
x = np.linspace(0, 20, 1000)
y1 = stats.gamma.pdf(x, a=3, scale=0.5)
y2 = stats.gamma.pdf(x, a=3, scale=1)
y3 = stats.gamma.pdf(x, a=3, scale=2)# plot
plt.figure(figsize=(15,7))
plt.plot(x, y1, label='λ=2')
plt.plot(x, y2, label='λ=1')
plt.plot(x, y3, label='λ=0.5')
plt.title('n=3, Waiting Till The 3rd Event', fontsize=24)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
plt.ylabel('PDF', fontsize=20)
plt.xlabel('Random Variable T', fontsize=20)
plt.show()
```

# Moment Generating Functions
Moments are simply the expected values of X
- E(X) : mean
- $E(X^2)$ : variance, uncentered, centered = $E(X^2) - \mu^2$
- $E(X^3)$ : Skewness = $E((X - \mu)^3) / \sigma^3
- $E(X^4)$ : Kurtosis = $E((X - \mu)^4) / \sigma^4
- $E(X^n)$: function that generates the moments
  - $MGF_x(t) = E[e^{tx}] = \sum e^{tx}.P(x) or \int e^{tx}.f(x)dx$
  - Derivate of MGF n times at t = 0 gives you E(X^n)
